```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier


diabete = pd.read_csv("diabetes.csv")
print(diabete.shape)


diabete = diabete[['Age', 'Glucose', 'Insulin', 'Outcome']]


Y = diabete[['Outcome']].values.ravel()
X = diabete.drop('Outcome', axis=1)


modele = KNeighborsClassifier()
modele.fit(X, Y)


def Consultation_diabete(Age, Glucose, Insulin, modele):

    Xn = pd.DataFrame([[Age, Glucose, Insulin]], columns=['Age', 'Glucose', 'Insulin'])
    prediction = modele.predict(Xn)
    proba = modele.predict_proba(Xn)[0]

    print("Prédiction:", prediction)
    print("Probabilité:", proba)

    labels = ['Non Diabétique (0)', 'Diabétique (1)']
    colors = ['black', 'orange']

    #plt.bar(labels, proba, color=colors)

    #plt.ylim(0, 1)
    #plt.ylabel("Résultat")
    #plt.xlabel("Probabilité")
    #plt.title(f"Probabilité de diabète pour les features suivants : Age={Age}, Glucose={Glucose}, Insulin={Insulin}")
    #plt.show()

    plt.pie(proba, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)


    plt.title(f"Probabilité de diabète pour les features suivants : Age={Age}, Glucose={Glucose}, Insulin={Insulin}")

    # Afficher le graphique
    plt.show()


Consultation_diabete(60, 70, 4, modele)
```
